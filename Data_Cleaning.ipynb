```python
from datetime import datetime

date_list = []
for item in user_data['yelping_since']:
    item = item.split("'")[1]
    date_list.append(item)
    
user_data['date'] = date_list
user_data['date'] = pd.to_datetime(user_data['date'])
print(datetime.today().strftime('%Y-%m-%d'))
print(type(datetime.today().strftime('%Y-%m-%d')))
today_datetime = datetime.strptime(datetime.today().strftime('%Y-%m-%d'), '%Y-%m-%d')


def calculatemonths(d1, d2):
    return (d2.year - d1.year) * 12 + d2.month - d1.month

months_list = []
for item in user_data['date']:
    months_list.append(calculatemonths(item, today_datetime))
print(months_list[:10])
print(user_data.head(10))
```
