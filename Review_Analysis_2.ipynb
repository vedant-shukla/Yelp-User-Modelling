```python
ranked_rev = pd.read_csv("review_with_score.csv")
```

```python
rev = ranked_rev[ranked_rev['user_id'] == user_1]
```

```python
user_ids = []
for user in ranked_rev['user_id']:
    user = user.split("'")[1]
    user_ids.append(user)

ranked_rev['user_id'] = user_ids


user_ids = []
for user in ranked_rev['business_id']:
    user = user.split("'")[1]
    user_ids.append(user)

ranked_rev['business_id'] = user_ids
```

```python
rev = ranked_rev[ranked_rev['user_id'] == user_1]
print(len(rev))
```

```python
rev.head(5)
```

```python
rev = rev.sort_values(['stars','sentiment_score'],ascending=False)
```
